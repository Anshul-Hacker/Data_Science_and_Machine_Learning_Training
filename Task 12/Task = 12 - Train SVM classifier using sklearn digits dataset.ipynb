```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import seaborn as sns
```

```python
digits = load_digits()
print(digits)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target'],axis='columns')
y = df.target
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.33,random_state=5)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
model = SVC(kernel='rbf',random_state=0, probability=True)
model.fit(x_train,y_train)
y_pred_1 = model.predict(x_test)
print("Model Score of Kernal (rbf) :", model.score(x_test,y_test))
```

```python
new_model=SVC(kernel='linear',random_state=0, probability=True)
new_model.fit(x_train,y_train)
y_pred_2 = new_model.predict(x_test)
print("Model Score of Kernal (linear) :", new_model.score(x_test,y_test))
```

```python
final_model = SVC(kernel='poly',random_state=0, probability=True)
final_model.fit(x_train,y_train)

y_pred_3 = final_model.predict(x_test)

print("Model Score of Kernal (poly) :", final_model.score(x_test,y_test))
```

```python
accuracy = accuracy_score(y_test,y_pred_3)
print('ACCURACY is',accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = np.array(confusion_matrix(y_test,y_pred_3))
cm
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred_3)
mse
```

```python
model_C=SVC(C=3)
model_C.fit(x_train,y_train)
model_C.score(x_test,y_test)
```

```python
new_model_C=SVC(C=3)
new_model_C.fit(x_train,y_train)
new_model_C.score(x_test,y_test)
```

```python
final_model_C=SVC(C=3)
final_model_C.fit(x_train,y_train)
final_model_C.score(x_test,y_test)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True, fmt=".2f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
A = f'Accuracy Score :{accuracy:.2f}'
plt.title(A)
plt.show()
```
