```python
# 10. Apply() functions

def rate(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'

df['Useless_col'] = df['Rating'].apply(rate)
df[['Rank','Title','Year','Rating','Votes','Metascore','Useless_col']]
```
